```python
import json
with open('../ar_wiki/data/units.json', 'r') as file:
    unit_data = json.load(file)

import pandas as pd
```

```python
data_cols = list()
for unit, unit_value in unit_data.items():
    print(unit_value['Element'])
    print(unit_value['Rarity'])
```

```python
name_list, rarity_list, passive_list, upgrades_list, element_list = [], [], [], [], []

for unit, unit_value in unit_data.items():
    name_list.append(unit_value['Name'])
    rarity_list.append(unit_value['Rarity'])
    passive_list.append(unit_value['Passives'])
    element_list.append(unit_value['Element'])
    upgrades_list.append(unit_value['Upgrades'])

df = pd.DataFrame({
    "Name": name_list,
    "Rarity": rarity_list,
    "Passives": passive_list,
    "Element": element_list,
    "Upgrades": upgrades_list
})
```

```python
df['Element'].unique()
```

```python
filtered = df[(df["Element"].isin(['Fighting', 'Dark'])) & (df["Rarity"] == "Secret")]

for element, group in filtered.groupby("Element"):
    print(f"--- {element} ---")
    print(group.Name)

total_secrets = df[df["Rarity"] == "Secret"]
secret_percentage = filtered.Name.count() / total_secrets.Name.count() * 100
```

```python
counts = total_secrets["Element"].value_counts()
all_elements = df['Element'].unique()

counts = counts.reindex(all_elements, fill_value=0)
print(counts)
```
