```python
import json
with open('../ar_wiki/data/units.json', 'r') as file:
    unit_data = json.load(file)

import pandas as pd
```

```python
data_cols = list()
for unit, unit_value in unit_data.items():
    for i in unit_value:
        data_cols.append(i)

data_df = pd.DataFrame(data_cols)
data_df[0].unique()
```

```python
name_list, rarity_list, passive_list, upgrades_list = [], [], [], []

for unit, unit_value in unit_data.items():
    name_list.append(unit_value['Name'])
    rarity_list.append(unit_value['Rarity'])
    passive_list.append(unit_value['Passives'])
    upgrades_list.append(unit_value['Upgrades'])

df = pd.DataFrame({
    "Name": name_list,
    "Rarity": rarity_list,
    "Passives": passive_list,
    "Upgrades": upgrades_list
})
```
